**16. Fournir un tableau qui indique le nombre de fois qu'un nom d'acteur se répète (sens de tri : nbr de répétition => ordre décroissant ?**

<div>
<img src="QUERY.png" width="400"/>
</div>

```python
SELECT COUNT(last_name)
FROM actor
GROUP BY last_name
HAVING COUNT(last_name) > 2
```

```python
mycursor.execute("SELECT last_name, COUNT(*) FROM actor GROUP BY last_name HAVING COUNT(*) > 2 ORDER BY COUNT(*) DESC")
for x in mycursor:
    print(x)
```

**17. Quel est le nom de famille qui se répète le plus ?**

```python
mycursor.execute("SELECT last_name, COUNT(*) FROM actor GROUP BY last_name HAVING COUNT(*) > 2 ORDER BY COUNT(*) DESC LIMIT 1")
for x in mycursor:
    print(x)
```

**18. Quels sont les `last_name` qui sont partagés par au moins deux acteurs dans la DB ?**


```python
select last_name, count(last_name) as 'Count of Last Name' 
from actor  group by last_name having count(last_name) > 2
```

```python
mycursor.execute("select last_name, count(last_name) as 'Count of Last Name' from actor group by last_name having count(last_name) > 2")
for x in mycursor:
    print(x)
```
